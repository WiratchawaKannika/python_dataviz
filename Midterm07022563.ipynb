```python
def  createDict(listnumber, listtext):
    if listnumber[0]%2 != 0 :
            numdict = {}  
            for index in range(len(listtext)):
                  numdict[listtext[index]] = listnumber[index] 
    else: 
        numdict = {}  
        for index in range(len(listtext)):
             numdict[listnumber[index]] = listtext[index]   
    return  numdict         
```

> จงหา Sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

- หา Sequence ที่มีขนาด 3 ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2):
        str_list.append(seq[i:i+3])
    return str_list
```

```python
 populate3charstring('ATTAAAGGT')
```

```python
list_all_3char = list()
for row in range(DNA.shape[0]):
    list_all_3char = list_all_3char + populate3charstring(DNA.iloc[row][0]) 
```

```python
unique_3char = list(set(list_all_3char))
```

```python
dict_3char = {}
for sub_str in unique_3char:
    num = 0
    for s in list_all_3char:
        if sub_str == s:
            num += 1
    dict_3char[sub_str] = num
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
     if dict_3char[k]>maxVal:
        maxVal = dict_3char[k]
        maxKey = k
```

```python
print(f'max 3-long-sequence is {maxKey} with frequency equals {maxVal}')
```

* สร้างตาราง Dictionary โดยตารางประกอบด้วย 2 Columns คือ ภาษาอีสานและภาษาไทย โดยค่าใน Columns ภาษาอีสานคือคำในภาษาอีสาน และ Columns ภาษาไทย คือคำแปลของภาษากลางของภาษาอีสานใน Columns แรก (ใช้ค่าจากตาราง data.xlsx โดยคำที่อยู่ที่อยู่ในลำดับเดียวกันของ columns ภาษาอีสาน และภาษาไทย คือคำแปลของกันและกัน ถ้าจำนวนคำของทั้ง 2 columns เท่ากัน -ถ้าจำนวนคำระหว่างสอง columns ไม่เท่ากันให้ข้าม recods นั้นไป)

```python
list_Esan = []
list_Thai =[]
for i in range(Data_Frame.shape[0]) :
    Esan_split = Data_Frame.iloc[i]['ภาษาอีสาน'].split(',')
    Thai_split = Data_Frame.iloc[i]['ภาษาไทย'].split(',')
    if len(Esan_split) == len(Thai_split):
        Match_split = len(Esan_split)
        for j in range(Match_split):
            list_Esan.append(Esan_split[j])
            list_Thai.append(Thai_split[j])
```
