## Scatter plot

```python
from matplotlib import pyplot as plt 
```

```python
import pandas as pd
```

```python
iris_data = pd.read_csv('C:\\Users\ASUS\\Documents\\Basic_Python\\python_dataviz\\iris.data' ,header=None)
```

```python
iris_data
```

* ต้องใช้ numpy เข้ามาช่วย

```python
import numpy as np
```

```python
setosa_np = iris_data.iloc[:50,:-1].to_numpy() #เอาตั้งแต่แถวที่ 0 ถึง 50 ตัดคอมลัมที่ 1 ออกไป
##setosa_np = iris_data.iloc[:50,:-3].to_numpy()
```

```python
setosa_np
```

> ใช้ Data 2 หลัก ในการ Plot 

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1]) #setosa_np[,]เอาแถวทุกแถวหลักที่ 0 ให้เป็ค่า X #setosa_np[:,1] เอาเอาทุกแถวหลักที่ 1 ให้เป็นค่า Y
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6) #s คือบอกขนาดของ marker  ว่าจะให้เป็นจุดเล็กหรือจุดใหญ่
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

> ใช้ Data 3 หลักในการ Plot

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2]) #เอาค่าหลักที่  3  มาเป็นขนาดของ marker
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2]) ##ขยายขนาดเพิ่มอีกด้วยการคูณได้ เพราะเป็นอาเรย์
```

> การใช้พารามิเตอร์กำหนดค่าสี

> วิธีที่ 1 กำหนดให้ marker เป็นสีเดียวกันทุกจุด

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'm') ##โดยใช้ตัวย่อของสี c สีฟ้าอ่อน,m สีม่วง,y สีเหลือง ,k สีดำ.r สีแดง,g สีเขียว,b สีน้ำเงิน
```

> วิธีที่ 2 กำหนดสีให้แต่ละตัว โดยที่แต่ละตัวสีต่างกัน

```python
setosa_np
```

> 1. หาค่า MAX กับ MIN  ของ setosa_np ก่อน

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

> 2. สร้าง list ให้กับ แต่ละช่วงตัวเลข ให้เปลี่ยนจากตัวเลขเป็นตัวอักษร String ตามสี 

```python
colors = []
for i in setosa_np[:,3] :
    if i < 0.15 :
        colors.append('b')
    elif i < 0.3 :
         colors.append('g')
    elif i < 0.45 :
         colors.append('y')
    else :
         colors.append('r')
colors
```

> 3. Plot โดยกำหนดค่าสี c =  List ที่สร้างไว้ ชื่อว่า colors

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2], c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2], c = setosa_np[:,3]) ## ลองให้ c เท่ากับแกนที่ 4 เป็นค่าสี
```

> วิธีที่ 3

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2], c = setosa_np[:,3], cmap='YlOrBr')
```

> วิธีที่ 4 

> แทนค่าสีด้วย RGB การผสมแม่สี 3 สี คือ g,r,b

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2], c = np.array([[1,0.5,0.8]]))กำหนดสีได้ตั้งแต่เลข 0 ถึง 1 สีเขียวเต็มๆ สีแดงนิดหน่อย สีม่วงมากหน่อย
```

> วิธีที่ 5

* รูปแบบของ marker

* สมารถเปลี่ยนรูปได้ ให้ทุกตัวหน้าตาเหมือนกัน ไม่สามารถทำให้แตกต่างกันได้

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2], c = setosa_np[:,3], cmap='PuBu',marker = '*')
```
