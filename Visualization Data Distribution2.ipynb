## Bubble Chart

```https://python-graph-gallery.com/272-map-a-color-to-bubble-plot/```

> Bubble Chart คือ การใช้ขนาดของวงกลมแสดงค่าของ Data 

````python
##plt.scatter(setosa_np[:,0],setosa_np[:,1],s=1000*setosa_np[:,2], c = 'r', alpha = 0.2)
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=3000*setosa_np[:,2], c=setosa_np[:,0], cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)
````

````python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=2000*setosa_np[:,2], c=setosa_np[:,0], cmap="bone", alpha=0.4, edgecolors="pink", linewidth=2,marker = '*')
````

````python
plt.scatter(virginica_np[:,3],virginica_np[:,0],s=2000*virginica_np[:,1], c = virginica_np[:,2], cmap="Blues",alpha=0.8, edgecolors="grey", linewidth=2)
# Add titles (main and on axis)
plt.xlabel("virginica is Features 4") ##ชื่อแกน X
plt.ylabel("Virginica is Features 1") ##ชื่อแกนY
plt.title("Virginica bubble plot") ##ชื่อ Bubble chart
plt.show()
````

- **alpha คือปรับสีให้ทึบหรือเข้มตามขนาดของตัวเลข มีค่าตั้งแต่ 0-1

## Map

> การ Plot Map การสำรวจภาวะเศรษฐกิจและสังคมของครัวเรือน ปี 2549 โดยใช้แผนที่ประเทศไทย

* import folium เพื่อใช้สำหรับการทำแผนที่

````python
import folium ##ต้องติดต้อง pip install folium
````

* นำ file Excel เข้ามา โดยตั้งชื่อเป็น 'Average Debt.xls'

````python
Debt = 'Average Debt.xls'
````

````python
Debt_df = pd.read_excel(Debt)
````

* ข้ามแถว 3 แถวบน ตัดออกเนื่องจากเป็นแถวที่เราไม่ต้องการ ใช้คำสั่ง Skiprows

````python
Debt_df = pd.read_excel(Debt,skiprows=3)
````

````python
Debt_df 
````

* เลือกเฉพาะ column ที่เราสนใจ

````python
Debt_df[['Region and province', 2549]]
````

````python
Debt_df_debt = Debt_df[['Region and province', 2549]]
````

* ลบแถวที่ไม่เกี่ยวข้องออกไป 

````python
Debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis =0,inplace=True)
````

````python
Debt_df_debt
````

````python
Debt_df_debt.head()
````

````python
data_to_plot = Debt_df_debt
````

````python
data_to_plot
````

````python
data_to_plot.head()
````

* Download ข้อมูลเชิงภูมิศาสตร์ของประเทศไทย จาก thai.json

````python
country_geo = 'thai.json'
````

* วน Loop ลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้

````python
new_col = list()
for d in data_to_plot['Region and province']:
     new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col
````

* กำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

````python
map = folium.Map(location=[10,100],  zoom_start =5)
````

````python
map
````

* ลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือน

* ถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมีค่ามากจะเป็นสีม่วงเข้ม

* จังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลืองตามแถบ

````python
map.choropleth(geo_data=country_geo,data=data_to_plot,
                          columns=['Region and province',2549],
                          key_on='feature.properties.name',
                          fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
                          legend_name='Average Debt')
````

* บันทึก Map ไว้บน HTML

````python
map.save('plot_data.html')
````
